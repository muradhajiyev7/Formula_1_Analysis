# **Dataset**
**Driver:** Name of the racing driver.

**Nationality:** The country to which the driver belongs.

**Seasons:** Number of seasons the driver has participated in.

**Championships:** Total number of championships won by the driver.

**Race_Entries:** Total number of races entered by the driver.

**Race_Starts:** Total number of races started by the driver.

**Pole_Positions:** Number of times the driver started a race from the pole position.

**Race_Wins:** Total number of races won by the driver.

**Podiums:** Total number of times the driver finished on the podium.

**Fastest_Laps:** Number of fastest laps achieved by the driver in races.

**Points:** Total points accumulated by the driver in their career.

**Active:** Indicates whether the driver is currently active or not.

**Championship Years:** Years in which the driver won championships.

**Decade:** The decade during which the driver had the most significant success or activity.

**Pole_Rate:** Percentage of races where the driver started from pole position.

**Start_Rate:** Percentage of races entered in which the driver started.

**Win_Rate:** Percentage of races won out of those entered.

**Podium_Rate:** Percentage of races finished on the podium out of those entered.

**FastLap_Rate:** Percentage of races where the driver achieved the fastest lap.

**Points_Per_Entry:** Average points earned per race entry.

**Years_Active:** Total years the driver has been active in the sport.

**Champion:** Indicates the years or seasons in which the driver became a champion.

```python
def heatmap(category):
    # Filter the DataFrame for only '2020s' decade
    df_2020 = df[df['Decade'] == '2020']
    
    # Pivot the data to create the heatmap structure
    heatmap_data = df_2020.pivot(index='Driver', columns='Decade', values='Points')

    plt.figure(figsize=(8, 6))  # Adjust figure size if needed

    sns.heatmap(heatmap_data, annot=True, cmap='viridis', fmt='g')

    plt.title('Points by Driver in 2020s')
    plt.xlabel('Decade')
    plt.ylabel('Driver')

    plt.show()
```

```python
df = pd.read_csv("F1DriversDataset.csv")

# Filter data for Decade = 2020
df_2020 = df[df['Decade'] == 2020]

# Pivot the filtered data
pivot_df = df_2020.pivot_table(index='Driver', columns='Decade', values='Points', aggfunc='sum', fill_value=0)

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_df, cmap='viridis', annot=True, fmt='g')
plt.xlabel('Decade')
plt.ylabel('Driver')
plt.title('Points by Driver in 2020')
plt.show()
```
